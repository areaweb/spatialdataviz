```python
from robobrowser import RoboBrowser
import geocoder
import folium
import pandas

browser = RoboBrowser(history=True, parser='html5lib')
url = 'http://www.t13.cl/noticia/nacional/te-puede-servir/ranking-colegios-mejor-promedio-psu-2016'
browser.open(url)
```

```python
table=browser.select('.article-content table')[0]
t=table.__str__()
t=t.replace(',','.')
df=pandas.read_html(t, header=0)[0]
df.head(100)
```

```python
for index,row in df.iterrows():
    g=geocoder.google(row.Colegio +", "+ row.Comuna + ", chile")
    if g:
        lat, lng = g.latlng
        df.set_value(index,'lat', lat)
        df.set_value(index,'lng', lng)
        print(row.Colegio, row.Comuna, g.latlng)      
```

```python
#df=df.drop('latlng',1).dropna()
df2 = df[df['Alumnos que rindieron la prueba']>30]
```

```python
map = folium.Map(location=[-33.39, -70.57], zoom_start=8, tiles='Stamen Toner')
i=0
for index, row in df2.iterrows():
    radio = 100
    map.add_child(
        folium.CircleMarker([row.lat,row.lng],
                            color='rgba(216, 67, 21, 0.500)',
                            fill_color='rgba(216, 67, 21, 0.500)', 
                            radius=4, 
                            popup=u"%s %s" % (row.Colegio, row['Promedio PSU'])))
    i+=1
map.save('colegios.html')
map
```
