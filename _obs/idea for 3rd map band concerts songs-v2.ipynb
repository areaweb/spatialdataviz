```python
from robobrowser import RoboBrowser
import geocoder
import folium
import pandas
import dateparser
from joblib import Memory
memory = Memory(cachedir='tmp', verbose=0)


@memory.cache
def get_latlng(query):
    g=geocoder.google(query)
    if g:
        print(query, g.latlng)
        return g.latlng
    else:
        return None, None
    
url = 'https://www.setlist.fm/search?page={}&query=Tool'
browser = RoboBrowser(history=True, parser='html5lib')
df = pandas.DataFrame(columns=['name', 'location', 'lat', 'lng', 'date'])
i=1

@memory.cache
def get_data_paged(query, page):
    url = 'https://www.setlist.fm/search?page={}&query={}'.format(page, query)
    browser.open(url.format(page))
    data = []
    for concert in browser.select('.setlistPreview'):
        month = concert.select('.month')[0].text
        day = concert.select('.day')[0].text
        year = concert.select('.year')[0].text
        datetext = "{}, {} {}".format(year, month, day)
        date = dateparser.parse(datetext)
        desc = concert.select('h2 a')[0].text
        idx = desc.find(' at ')+4
        loc = desc[idx:]
        loc_pieces = loc.split(',')
        if loc == None:
            continue
        if len(loc_pieces)>=3:
            loc = ','.join(loc_pieces[-3:])
        lat, lng = get_latlng(loc)
        songs = concert.select('.list-inline li')
        for song in songs:
            songname = song.text
            if songname != '...':
                data.append([songname, loc, lat, lng, date])
    return data
    
i=1
for page in range(1,30):
    print(page)
    data = get_data_paged('Tool', page)
    for datum in data:
        df.loc[i] = datum
        i+=1
df=df.dropna()
df
```

```python
top_songs = df.name.value_counts().head(6)
colors = ['red','blue','yellow','green','orange','purple']
map = folium.Map(zoom_start=8, tiles='OpenStreetMap')
i=0
for song in top_songs.keys():
    i+=1
    color = colors.pop()
    dfc = df[df['name']==song]
    for index, row in dfc.iterrows():
        folium.Marker(
            location=[row.lat+i/5.0,row.lng+i/5.0],
            #popup=u"%s - %s - %s" % (row['name'], row['location'], row['date']),
            icon=folium.Icon(color=color)
        ).add_to(map)
map
```
