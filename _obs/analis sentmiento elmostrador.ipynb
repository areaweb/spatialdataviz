```python
from __future__ import unicode_literals
from robobrowser import RoboBrowser
from textblob import TextBlob
from collections import Counter

urls = 'http://www.elmostrador.cl/','http://www.elmostrador.cl/noticias/pais','http://www.elmostrador.cl/noticias/mundo','http://www.elmostrador.cl/noticias/opinion','%%!'

print 'starting'
browser = RoboBrowser(history=True, parser='html5lib')


        

```

```python
titulos_vistos = []

from collections import Counter
from pattern.es import parse, split

for url in urls:
    browser.open(url)
    titulos = browser.select('h3 a') + browser.select('h2 a')

    for titulo in titulos:
        try:
            browser.follow_link(titulo)
            texto = browser.select('.cuerpo-noticia')[0].text
            titulo = browser.select('.datos-noticias h2')[0].text
            if titulo in titulos_vistos:
                continue
            titulos.append(titulo)
            blob = TextBlob(texto)
            if blob.sentiment.polarity:
                print u"%s" % titulo, str(blob.sentiment.polarity), str(blob.sentiment.subjectivity)
            c=Counter()
            #print blob.tags
            for word, tag in blob.tags:
                print "WT:", word, tag
                if tag == u'NN':
                    w=word.lemmatize()
                    print "NN: ", w
                    c[w]+=1
            print c
        except IndexError:
            pass
```
