```python
from robobrowser import RoboBrowser
import plotly.plotly as py
import pandas as pd

browser = RoboBrowser(history=True, parser='html5lib')
base_url = 'https://flightaware.com/live/fleet/'
query = 'AAL' ## LAN KLM JAL BAW AAL CGW CCA
df = pd.DataFrame()
for i in range(10):
    url = base_url + query + "?;offset=" + str(i*20)
    browser.open(url)
    table=browser.select('.prettyTable')[0]
    t=table.__str__()
    if 'Sorry' in t:
            break
    tdf=pd.read_html(t, header=1)[0]
    df = df.append(tdf)
print(i)
df
```

```python
from cachier import cachier
import geocoder


@cachier()
def get_latlng(query):
    g=geocoder.google(query)
    if g:
        return g.latlng
    else:
        return None, None
    
great_lines = []
airports = []
airport_data = []
seen=[]
great_lines_2=[]
airports = pd.DataFrame(columns=['label','lat','lon'])
i=0
for index,row in df.iterrows():
    origin = row.Origin
    destination = row.Destination
    olat, olon = get_latlng(origin)
    dlat, dlon = get_latlng(destination)
    if olat and dlat:
    
        if origin not in seen:
            seen.append(origin)
            #print(origin, olat, olon)
        if destination not in seen:
            seen.append(destination)
            #print(destination, dlat, dlon)
        great_lines.append(dict(
        type = 'scattergeo',
        lat = [ olat, dlat ],
        lon = [ olon, dlon ],
        mode = 'lines',
        line = dict(
            width = 2,
            color = 'red',
            ),
            )
        )
        airports.loc[i]=[origin, olat, olon]
        airports.loc[i+1]= [destination, dlat, dlon]
        i+=2
airport_markers = [ dict(
        type = 'scattergeo',
        lon = airports['lon'],
        lat = airports['lat'],
        hoverinfo = 'label',
        text = airports['label'],
        mode = 'markers',
        marker = dict( 
            size=8, 
            color='red',
            line = dict(
                width=3,
                color='rgba(68, 68, 68, 0)'
            )
        ))]
    
layout = dict(
        title = 'Great lines over globe<br>(Click and drag to rotate)',
        width = 800,
        height = 800,
        showlegend = False,         
        geo = dict(
             showland = True,
            landcolor = '#f9f9f9',
            showocean = True,
            countrywidth = 0.5,
            #landcolor = 'rgb(230, 145, 56)',
            lakecolor = 'rgb(240, 255, 255)',
            oceancolor = 'rgb(240, 255, 255)',
            projection = dict( 
                type = 'orthographic',
                rotation = dict(
                    lon = 0,
                    lat = 0,
                    roll = 0
                )            
            ),
            lonaxis = dict( 
                showgrid = True,
                gridcolor = 'rgb(102, 102, 102)',
                gridwidth = 0.5
            ),
            lataxis = dict( 
                showgrid = True,
                gridcolor = 'rgb(102, 102, 102)',
                gridwidth = 0.5
            )
        )
    )
    
fig = dict( data=great_lines+airport_markers, layout=layout )
py.iplot( fig, validate=False, filename='d3-globe' )
```
